```python
from iterstrat.ml_stratifiers import MultilabelStratifiedKFold
```

```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv("Customer-Attrition-Prediction-main/train_encoded.csv")
train_df.head(5)
```

```python
folds = train_df.copy()
NFOLDS = 9 #Splits
mskf = MultilabelStratifiedKFold(n_splits=NFOLDS)

ms_cols = np.array(train_df[['CustomerAttrition','CyberProtection','MobileService','Aged','TotalDependents','SettlementProcess_Bank','FilmSubscription','SettlementProcess_Card','SettlementProcess_Check','SettlementProcess_Electronic']])

for f, (t_idx, v_idx) in enumerate(mskf.split(X=train_df, y=ms_cols)):
    folds.loc[v_idx, 'kfold'] = int(f)

folds['kfold'] = folds['kfold'].astype(int)
folds
```

```python
folds.to_csv("train_encoded_folds.csv",index=False)
```
